```python
import os
import json
import pickle
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
from collections import OrderedDict
```

```python
BASE_LOG_DIR = "../log/grad/"
```

```python
history_path = glob(os.path.join(BASE_LOG_DIR, "*.json"))
```

```python
history = OrderedDict()
```

```python
for path in history_path:
    print(path)
    day = path.split("/")[2]

    with open(path, "r") as f:
        history.update({day: json.load(f)})
```

```python
legend = []
for key in history.keys():
    print("{0} Max Val Acc: ".format(key), max(history[key]["val_acc"])*100)
    plt.plot(history[key]["acc"])
    plt.plot(history[key]["val_acc"])
    legend.extend(["acc_{0}".format(key), "val_acc_{0}".format(key)])
plt.grid()
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(legend)
plt.show()
plt.close()

legend = []
for key in history.keys():
    plt.plot(history[key]["loss"])
    plt.plot(history[key]["val_loss"])
    legend.extend(["loss_{0}".format(key), "val_loss_{0}".format(key)])
plt.grid()
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(legend)
plt.show()
plt.close()
```
