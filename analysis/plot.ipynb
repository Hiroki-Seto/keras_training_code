```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

from glob import glob
from collections import OrderedDict
```

```python
BASE_LOG_DIR = "../log/hyperparam/"
```

```python
history_path = glob(os.path.join(BASE_LOG_DIR, "*/*.json"))
```

```python
bs_ex_path = [path for path in history_path if "base" in path] + [path for path in history_path if "bs_" in path]
```

```python
lr_ex_path = [path for path in history_path if "base" in path] + [path for path in history_path if "lr_" in path]
```

```python
bs_history = OrderedDict()

for path in bs_ex_path:
    print(path)
    day = path.split("/")[3]

    with open(path, "r") as f:
        bs_history.update({day: json.load(f)})
    print("{0}: ".format(day), max(bs_history[day]["val_acc"])*100)
```

```python
lr_history = OrderedDict()

for path in lr_ex_path:
    print(path)
    day = path.split("/")[3]

    with open(path, "r") as f:
        lr_history.update({day: json.load(f)})
    print("{0}: ".format(day), max(lr_history[day]["val_acc"])*100)
```

```python
legend = []
for key in bs_history.keys():
    print("{0} Max Val Acc: ".format(key), max(bs_history[key]["val_acc"])*100)
    plt.plot(bs_history[key]["acc"])
    plt.plot(bs_history[key]["val_acc"])
    legend.extend(["acc_{0}".format(key), "val_acc_{0}".format(key)])
plt.grid()
plt.legend(legend)
plt.ylabel('acc')
plt.xlabel('epoch')
plt.show()
plt.close()

legend = []
for key in bs_history.keys():
    plt.plot(bs_history[key]["loss"])
    plt.plot(bs_history[key]["val_loss"])
    legend.extend(["loss_{0}".format(key), "val_loss_{0}".format(key)])
plt.grid()
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(legend)
plt.show()
plt.close()
```

```python
legend = []
for key in lr_history.keys():
    print("{0} Max Val Acc: ".format(key), max(lr_history[key]["val_acc"])*100)
    plt.plot(lr_history[key]["acc"])
    plt.plot(lr_history[key]["val_acc"])
    legend.extend(["acc_{0}".format(key), "val_acc_{0}".format(key)])
plt.grid()
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(legend)
plt.show()
plt.close()

legend = []
for key in lr_history.keys():
    plt.plot(lr_history[key]["loss"])
    plt.plot(lr_history[key]["val_loss"])
    legend.extend(["loss_{0}".format(key), "val_loss_{0}".format(key)])
plt.grid()
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(legend)
plt.show()
plt.close()
```
